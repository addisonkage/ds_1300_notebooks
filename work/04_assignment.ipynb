```python
import pandas as pd
import numpy as np

%config InlineBackend.figure_format ='retina'
```

```python
per=pd.read_csv('/users/akage/ds_1300_notebooks/project/PER_Ranking copy.csv')
```

```python
Names=per['Player'].str.split(pat='\\', expand= True)
Names['Rk']=per['Rk']
Names.columns=['Player_','ID','Rk']
Named=Names.reindex(columns=['Rk','Player','ID'])
Names.drop('ID',axis=1,inplace=True)
Names.head()
```

```python
Names['FirstInit'] = Names['Player_'].apply(lambda x: x[0])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
    

p1=plt.scatter(Names['FirstInit'], Names['Rk'],cmap='coolwarm',c=Names['Rk'],edgecolor='black')
  
# Adding Title to the Plot
plt.title("Rank by Initial")
  
# Setting the X and Y labels
plt.xlabel('FirstInit')
plt.ylabel('Rk')

plt.colorbar()

#initialrank=plt.show()
```

```python
p1.figure.savefig('figure.pdf')
```

This plot shows the first initial of a player in comparison to their rank. 
The graph goes from most values at the left to least values at the right. 
This graph includes all the players in our table and demonstrates what the "best" first initial is
in regards to NBA rank. 

I chose to make the plot the way that I did because I wanted to visualize not only how many 
players with a certain initial there are in the top NBA players but also what the distribution of rank
looks like. Additionally, I chose the color scheme to represent the NBA colors, but also outlined
all the data points to make sure none were lost with the white of the color scheme.
